```python
# 下载并保存 resnet18 到 ./pre_model 中
import os
import torch
from torchvision import models

os.makedirs('./pre_model', exist_ok=True)

# 下载预训练的 ResNet-18（会从网络获取权重）
model = models.resnet18(pretrained=True)

# 保存 state_dict 到指定目录
save_path = './pre_model/resnet18.pth'
torch.save(model.state_dict(), save_path)
print(f"ResNet-18 已保存到: {save_path}")
```

```python
# 替换原来调用 torchvision.datasets.BloodMNIST 的单元格，改用 medmnist
from medmnist import BloodMNIST
import torchvision.transforms as transforms
import os

os.makedirs('./datasets/bloodmnist', exist_ok=True)
transform = transforms.Compose([transforms.ToTensor()])

# 不自动下载，要求手动把 npz 放到 root 中
npz_path = './datasets/bloodmnist/bloodmnist.npz'
if not os.path.exists(npz_path):
    raise RuntimeError(
        "找不到 bloodmnist.npz，请在终端运行以下命令下载后重试:\n"
        "curl -L -o ./datasets/bloodmnist/bloodmnist.npz \"https://zenodo.org/records/10519652/files/bloodmnist.npz?download=1\"\n"
        "下载后可通过 `md5sum bloodmnist.npz` 验证 MD5: 7053d0359d879ad8a5505303e11de1dc"
    )

bloodmnist_train = BloodMNIST(root='./datasets/bloodmnist', split='train', download=False, transform=transform)
bloodmnist_test  = BloodMNIST(root='./datasets/bloodmnist', split='test',  download=False, transform=transform)

print("BloodMNIST 数据集已加载（已存在文件）")
```
